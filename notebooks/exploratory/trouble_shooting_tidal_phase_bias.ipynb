```python
import matplotlib.pyplot as plt

atg_M2_phases = []
roms_M2_phases = []
atg_O1_phases = []
roms_O1_phases = []
for station,data in sta.items():
    atg_M2_phases.append(data['atg']["M2"][1])
    roms_M2_phases.append(data['tt']['M2'][2])
    atg_O1_phases.append(data['atg']["O1"][1])
    roms_O1_phases.append(data['tt']['O1'][2])
```

```python
plt.close()
plt.plot(atg_M2_phases,'r.',label='atg')
plt.plot(roms_M2_phases,'b.',label='roms')
plt.show()
plt.plot(atg_O1_phases,'r.',label='atg')
plt.plot(roms_O1_phases,'b.',label='roms')
plt.legend()
plt.show()
```

```python
len(roms_M2_phases) == len(atg_M2_phases)
```

```python
diff_M2 = np.array(roms_M2_phases) - np.array(atg_M2_phases)
diff_O1 = np.array(roms_O1_phases) - np.array(atg_O1_phases)
plt.close()
plt.title('Phase difference at ATG stations (ROMS - ATG) in deg (upper M2, lower O1)',fontsize=16)
for diff in [diff_M2,diff_O1]:
    diff[diff>180.0]-=360
    diff[diff<-180.0]+=360
    plt.plot(diff,'r.')
    plt.xlabel('station')
    plt.ylabel('roms phase lag to atg in deg')
    plt.show()
    print('mean:',np.nanmean(diff),'; std: ',np.nanstd(diff),'; rms: ',np.sqrt(np.nanmean(diff**2)))
```
